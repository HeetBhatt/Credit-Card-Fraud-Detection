```python
#let's import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importint dataset
import os
os.chdir('D:\python\creditcardfraud\creditcard.csv')
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
# now let's clean the data
df.groupby('Time')['Time'].agg('count')
```

```python
# let's find null values
df.isnull().sum()
```

```python
# there is not any null value in our dataset
# now let's find the duplicates
duplicated_rows=df[df.duplicated()]
duplicated_rows
```

```python
# there are many duplicated rows
#we have to remove that data

df.drop_duplicates(inplace=True)
```

```python
#now let's see the shape of our dataset
df.shape
```

```python
#now let's explore the target variable
df.groupby('Class')['Class'].agg('count')
```

```python
sns.countplot(x='Class',data=df)
```

```python
# our data is very biased
sns.boxplot(x='Class',y='Amount',data=df)
```

```python
# we have to handle this biased variable. One way to do this is called undersampling.
```

```python
from imblearn.under_sampling import NearMiss
```

```python
# we have to divide the data in two parts that is dependent and independent data
Columns=df.columns.tolist()
Columns.remove('Class')

Columns

```

```python

x=df[Columns]
y=df['Class']
print(x.shape)
print(y.shape)
```

```python
#implementing undersampling
nm=NearMiss()
x_res,y_res=nm.fit_sample(x,y)
```

```python
print(x_res.shape)
print(y_res.shape)
```

```python
# in undersampling we have just take 473 records from class=0 data, so that it is equal to number of records of class=1 data
from collections import Counter
print('original dataset shape {}'.format(Counter(y)))
print('after undersampling dataset shape {}'.format(Counter(y_res)))
```

```python
data=pd.concat([x_res,y_res],axis=1)
data
```

```python
sns.boxplot(x='Class',y='Amount',data=data)
```

```python
sns.countplot(x='Class',data=data)
```

```python
#now we are going to remove outliers from dataset
max_thresold=df['Amount'].quantile(0.95)
min_thresold=df['Amount'].quantile(0.05)
```

```python
#outliers from down side
mindata=data[data['Amount']<min_thresold]
mindata
```

```python
# let's start model buoilding
# we have to devide the data into two parts for fraud and normal
fraud=data[data['Class']==1]
normal=data[data['Class']==0]

```

```python
# for model building let's split our dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x_res,y_res,test_size=0.2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
# we have to use logistic regression to classify the data
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# model evaluation
# accuracy on training data

x_train_prediction=model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
training_data_accuracy=accuracy_score(x_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(x_test_prediction,Y_test)
print(testing_data_accuracy)
```
